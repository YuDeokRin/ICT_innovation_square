```
import pandas as pd
```

```
df01=pd.DataFrame({
    "eng_score":[1, 4, 2, 9, 9],
    "kor_score":[3, 2, 7, 3, 10]
})
```

```
df01
```

```
#영어 점수 평균
df01["eng_score"].mean()
```

```
#국어 점수 평균
df01["kor_score"].mean()
```

```
#df01["eng_score"]:영어 점수

#df01["eng_score"].name() : #영어 점수 평균
df01["eng_score"] - df01["eng_score"].mean()
```

```
#df01["kor_score"]:국어 점수

#df01["kor_score"].name() : #국어 점수 평균
df01["kor_score"] - df01["kor_score"].mean()
```

```
(df01["eng_score"] - df01["eng_score"].mean()) * (df01["kor_score"] - df01["kor_score"].mean())
```

```
((df01["eng_score"] - df01["eng_score"].mean()) *( (df01["kor_score"] - df01["kor_score"].mean()) ) ).sum()
```

```
((df01["eng_score"] - df01["eng_score"].mean()) *( (df01["kor_score"] - df01["kor_score"].mean()) ) ).sum()/4
```

```
df01.cov()  #kor_score :11.50 (국어편차)^2/4 = 분산   4.24 = 공분산  
            #eng_score : 14.50 (영어편차)^2/4 = 분산  4.24  = 공분산    
```

```
#영어 점수의 표준편차
df01["eng_score"].std()
```

```
#국어 점수의 표준편차
df01["kor_score"].std()
```

image.png

```
#영어 국어 점수의 공분산
df01.cov()/(df01["eng_score"].std() * df01["kor_score"].std())
```

```
#상관 관계 계산
df01.corr()
```
