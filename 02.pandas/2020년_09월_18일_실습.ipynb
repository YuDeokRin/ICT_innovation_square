5명 사원의 월급이 저장된 DataFrame 생성

```
sal_df = pd.DataFrame(
    {"sal":[200,300,280,500,470]},
    index=["길동","길순","태희","지훈","효리"]
    
)
```

```
sal_df
```

```
#이상치가 없을때 평균 => 대표값
sal_df["sal"].mean()
```

>워렌버핏이 자기 월급도 분석해 달라고 함
>워렌버핏의 월급 300억 (평범한 경우가 아니라 매우 희귀하고 특별한 경우 -> 이상치)


이상치 (워렌버핏)을 sal_df에 추가해서 평균을 구해보기

```
sal_df.loc["워렌버핏"]=3000000
```

```
sal_df
```

```
#이상치를 추가 한 후에 월급 평균 : 50억
sal_df["sal"].mean()
```

 > 이상치가 있으면 평균은 값을 대표가 어려움
 
 > 50억이 사원의 월급을 대표하기 어려움
 
 > 이상치가 있을때 컬럼의 값을 하나로 요약하고 싶다 => 중간값

```
#이상치가 없을 때 중앙값(평균과 비슷)
sal_df["sal"].median()
```

```
sal_df["sal"]
```

```
#중앙값 계산
sal_df["sal"].median()
```

```
#df 데이터 프레임의 각 칸의 중간값
df.median()
```

```
df["mpg"].median()
```

각 컬럼의 최대값, 최소값 => 근처에 이상치가 있음

```
#sal 컬럼에 이상치가 있다.(이상치 매우 큰값)
sal_df["sal"].max()
```

```
#sal 컬럼에 매우 작은 이상치는 없다. 
sal_df["sal"].min()
```
