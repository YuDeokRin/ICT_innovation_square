```python
import numpy as np
```

```python
#X는 독립변수로 입력값
X=np.array([1,2,3], dtype="float32")
#Y는 종속 변수로 출력값
y= np.array([2, 2.5, 3.5],dtype="float32")
```

```python
X
```

```python
y
```

```python
#초기 선형 회귀 식은 wX+b = 2X+1
w = 2
b = 1
```

```python
#X*W+b를 실행해서 예측값을 hypothesis에 대입
hypothesis = X*w+b
hypothesis
```

```python
#Gradient Decent 편미분 공식에 의해서 w_gred계산
w_gred = 1/3*np.dot((hypothesis-y),X.T)
w_gred
```

```python
#w 업데이트
w = w - w_gred
w
```

```python
#Gradient Decent 편미분 공식에 의해서 b_gred계산
b_gred= 1/3 * np.sum(hypothesis-y)
b_gred
```

```python
#b 업데이트
b = b -b_gred
b
```

```python
#Gradient Decent 알고리즘을 반복 적으로 실행해서 오차제곱의 합이 최소화 되는 w와 b를 찾음

for i in range(100):
    #예측값을 hypothesis에 대입
    hypothesis = X*w+b
    print("hypothesis:",hypothesis)
    #예측값-진짜값의 제곱의 합 coast를 계산
    MSE=np.sum((hypothesis-y)**2)
    print("MSE:",MSE)
    #w의 수정값
    w_gred = 1/3*np.dot((hypothesis-y),X.T)
    print("w_gred:",w_gred)
    #b의 수정값
    b_gred = 1/3 * np.sum(hypothesis-y)
    print("b_gred:",b_gred)
    #w업데이트
    w=w-w_gred
    print("w:",w)
    #b업데이트
    b=b-b_gred
    print("b:",b)
    
    
```

```python
#w의 초기깂
w = 2
#b의 초기값
b = 1
#learninng rate
learning_rate=0.1

#100 번 반복해서 오차 제곱의 합이 0으로 수렴되는 w와 b를 찾음
for i in range(1000):
    #예측값을 hypothesis에 대입
    hypothesis = X*w+b
    print("hypothesis:",hypothesis)
    #예측값-진짜값의 제곱의 합 coast를 계산
    MSE=np.sum((hypothesis-y)**2)
    print("MSE:",MSE)
    #w의 수정값
    w_gred = 1/3*np.dot((hypothesis-y),X.T)
    print("w_gred:",w_gred)
    #b의 수정값
    b_gred = 1/3 * np.sum(hypothesis-y)
    print("b_gred:",b_gred)
    #w업데이트
    w=w-(learning_rate*w_gred)
    print("w:",w)
    #b업데이트
    b=b-(learning_rate*b_gred)
    print("b:",b)
```

```python
#X에 곱해지는 값 w를 출력
print("w=",w)
#X에 더해지는 값 b를 출력
print("b=",b)
```
