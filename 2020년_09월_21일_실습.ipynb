```
df=pd.DataFrame({"x":[1,2,3,4,5],
                 "y":[0.2, 0.7, 0.8, 0.9,2.5]
                 })
```

```
df
```

```
#컬럼의 평균
df.mean()
```

```
#x컬럼
df["x"]
```

```
#x컬럼 평균
df["x"].mean()
```

```
#x컬럼-x컬럼 평균
df["x"]-df["x"].mean()
```

```
#y컬럼-y컬럼 평균
df["y"]-df["y"].mean()
```

image.png

```
#(x컬럼-x컬럼 평균) * (y컬럼-y컬럼 평균)
(df["x"]-df["x"].mean()) * (df["y"]-df["y"].mean())
```

```
#( (x컬럼-x컬럼 평균) * (y컬럼-y컬럼 평균) ).sum() :합
( (df["x"]-df["x"].mean()) * (df["y"]-df["y"].mean()) ) .sum()
```

image.png

```
#공분산
#( (x컬럼-x컬럼 평균) * (y컬럼-y컬럼 평균) ).sum() /4
( (df["x"]-df["x"].mean()) * (df["y"]-df["y"].mean()) ) .sum()/4
```
