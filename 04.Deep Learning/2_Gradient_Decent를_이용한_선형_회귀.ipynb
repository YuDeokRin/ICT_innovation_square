```
w=2
b=1
learning_rate=0.1

for i in range(100):
  print("w =>", w)
  print("b =>", b)
  hypothesis = X*w+b
  print("hypothesis =>", hypothesis)
  cost=np.sum((hypothesis-y)**2)
  print("cost=>",cost)
  w_gred=1/3*np.dot((hypothesis-y),X)
  print("w_gred=>",w_gred)
  b_gred=1/3*np.sum(hypothesis-y)
  print("b_gred=>",b_gred)
  w=w - (learning_rate* w_gred)
  b=b - (learning_rate*b_gred)
  print("="*100)
```
