```
titanic.loc[: , ["age","sex","class","fare","survived"]]
```

```
df=titanic.loc[: , ["age","sex","class","fare","survived"]]
```

```
df
```

image.png

```
pd.pivot_table(
    df,
    index="class",
    columns="sex",
    values="age",
    aggfunc="mean"
)
```

image.png

```
pd.pivot_table(
    df,
    index="class",
    columns="sex",
    values="age",
    aggfunc=["mean","sum"]
)
```

image.png

```
pd.pivot_table(
    df,
    index=["class","sex"],
    columns="survived",
    values=["age","fare"],
    aggfunc=["mean","max"]
)
```
